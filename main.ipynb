```python
df = pd.merge(df,cc_df,on='Country Code',how='left')
```

```python
df['Country Code'].unique()
```

```python
df['Country'].value_counts()
```

```python
plt.pie(df['Country'].value_counts().values[:3], labels=df['Country'].value_counts().index[:3],autopct='%1.2f%%');
```

Zomato max records of transactions are from India, followed by United States and United Kingdom

```python
df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

here is how zomato assigns the colors and rating texts according to aggregate rating (basically it's their scale of rating)

```python
rating_df = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
rating_df
```

```python
plt.figure(figsize=(14, 8),dpi=200)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=rating_df,palette=['gray','red','orange','yellow','lightgreen','green']);
```

I set gray color for white as otherwise u can't see shit

The highest counts got the worst ratings, and it has almost normal distribution (bell curve) without taking into account the unrated ones
